```python
x = melb_data['Lattitude']
y = melb_data['Longtitude']
sns.scatterplot(x=x, y=y)
```

```python
suburb_count = melb_data.Suburb.value_counts()
suburb_count = suburb_count[:10,]

sns.set(style='darkgrid')
plt.figure(figsize=(18, 6))
plt.xticks(rotation=0)
plt.title('houses sold in top 10 Suburb in the melbourne')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Suburb', fontsize=12)

sns.barplot(x=suburb_count.index, y=suburb_count.values, alpha=0.8, )
```

```python
suburb_count = melb_data.Suburb.value_counts()
suburb_count = suburb_count[:50,]

sns.set(style='darkgrid')
plt.figure(figsize=(18, 6))
plt.xticks(rotation=90)
plt.title('houses sold in Suburb in the melbourne')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Suburb', fontsize=12)

sns.barplot(x=suburb_count.index, y=suburb_count.values, alpha=0.8, )
```
